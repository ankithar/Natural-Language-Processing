```python
C=4
logPrior=list()
logLiklihood = collections.defaultdict(list)
ndoc = lengthPosTruth+lengthNegTruth+lengthPosDec+lengthNegDec
def train_NB():
    for i in range(C):
        if(i==0):
            bigDoc = countsPosTruth
            nc = lengthPosTruth
        if(i==1):
            bigDoc = countsPosDec
            nc = lengthPosDec
        if(i==2):
            bigDoc = countsNegTruth
            nc = lengthNegTruth
        if(i==3):
            bigDoc = countsNegDec
            nc = lengthNegDec
            
        logPriorValue = nc/ndoc
        logPrior.append(np.log(logPriorValue))
        
        for word in vocab.keys():
            countWC = bigDoc[word]
            countWPC = vocab[word]
            #print(countWPC)
            prob = (countWC+1)/(countWPC+len(vocab))
            #prob = (countWC+1)/(countWPC)
            logProb = np.log(prob)
            logLiklihood[word].append(logProb)
```
